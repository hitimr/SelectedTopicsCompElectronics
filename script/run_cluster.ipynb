```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime

import common
```

```python
def to_str_args(args):
    str_args = ""
    for key in args:
        if args[key] == "":
            str_args += f"--{key} "
        else:
            str_args += f"--{key} {args[key]} "
    return str_args



# Cluster specific options
timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
gpu_short_name = "Tesla_T4"
cpu_short_name = "Xeon_6248"
outfile = common.globals["paths"]['out_dir'] + f"{timestamp}_{cpu_short_name}_{'gpu_short_name'}.csv"

args = {
    "outfile" : outfile,
    "p_rays_start": 6,
    "p_rays_end": 26,
    "n_bench": 4,
    "cpu_price": 3290.0,
    "cpu_power": 150.0,
    "omp_n_threads": 40,
    "gpu_price": 2500.0,
    "gpu_power": 70.0,
    "gpu_grid_size": 256,
    "gpu_block_size": 256,
    "skip-openvdb" : "",
    "skip-nanovdb-cpu": "",
    "skip-checks": ""
}




str_args = to_str_args(args) 
str_args

common.run_benchmark(outfile, str_args, print_output=True, verbose=True)
```
